In order to baseline the house prices, they have been adjusted using a pricing index found 
https://www.thebalance.com/u-s-inflation-rate-history-by-year-and-forecast-3306093
What we did was setting 2003 an as baseline (index 100) - this mean that 1$ in 2003 was worth 1$ (because 1$/100% = 1$), and then we simply multiply the pricing index for each year with all the previous years giving us the following result. 
Below is a table showing the accumulated price index running from year 2003 to 2017
values have been rounded.

```python
#year    priceindex    1$ worth 

#2003    1.00        1.00
#2004    1.03        0.97        
#2005    1.06        0.94    
#2006    1.09        0.92
#2007    1.13        0.88
#2008    1.13        0.88        
#2009    1.16        0.86
#2010    1.17        0.85
#2011    1.21        0.83
#2012    1.23        0.81
#2013    1.25        0.80
#2014    1.26        0.79
#2015    1.27        0.79
#2016    1.29        0.78
#2017    1.32        0.76

#(code cell to keep table format)
```

This evens out the purchasing power between all the years in our dataset - in effect making it s that each of the datapoints were recorded in 2003.

### Building a neural network model 

Splitting the dataset into a training part and a test part is done partly to prevent overfitting - if we train our model on 100% of the data we end up with a model almost exclusively cast to that data, which obviously makes it less usefull on any other data. 

And partly to have some data to test / verify the models accuracy on after having trained it - the test part is used to verify the model. Typically the ratio of as such a split is 80 / 20 (training/testing)

Using adjusted price as label/target and the remaining features as data, we divide the dataset into training and testing. 

```python
x = df.iloc[:,:-1].values
y = df['adjusted_price'].values

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=5)
```

We create a squential neural network with 4 hidden layers containing 30 neurons each. The model has 9 input neurons, since we are interested in only one single output (building price) the model only has a single output neuron

```python
model = Sequential()
model.add(Dense(9, activation='relu'))
model.add(Dense(30, activation='relu'))
model.add(Dense(30, activation='relu'))
model.add(Dense(30, activation='relu'))
model.add(Dense(30, activation='relu'))
model.add(Dense(1))
```

Compiling the model

```python
model.compile(loss='mean_squared_error', optimizer='adam')
```

Training the model, and storing the output in the history variable

```python
history = model.fit(x_train, y_train, epochs = 10)
```

After training our model we evaluate loss. This is a very large loss. With a loss of this magnitude our model is practically useless.

```python
model.evaluate(x_test, y_test)
```

Using from the history variable we stored our model we plot the loss. This gives us the following diagram depicted in the cell below. We see that the magnitude of the loss decreasing for each epoch, meaning that ultimately if we were to train the model for many epochs it is likely that we end up with a usable model (acceptable loss magnitude)

```python
plt.plot(history.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train'], loc='upper left')
plt.show()
```
